# Extract the target variable.


```python
final_data = churn[['tenure','MonthlyCharges','SeniorCitizen','Churn']]
final_data
```

```python
final_data.dtypes
```

```python

def converttype(x):
    if x == 'No':
        return 0
    elif x == 'Yes':
        return 1
    else:
        return x
    
final_data['Churn'] = final_data['Churn'].apply(converttype)
final_data.head()
```

```python
final_data['Churn'] = final_data['Churn'].astype(int)
final_data.head()
```

```python
final_data.dtypes
```

```python
corr_matrix=final_data.corr(method='pearson')  # default
fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
y = final_data['Churn']
X = final_data.drop('Churn', axis =1)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Selecting the numeric columns
X_train_num = X_train.select_dtypes(include = np.number)

# Applying transformer on numeric train data
transformer = MinMaxScaler().fit(X_train_num)
X_train_normalized = transformer.transform(X_train_num)
X_train_normalized = pd.DataFrame(X_train_normalized,columns=X_train_num.columns)
X_train_normalized.head()
```

```python

from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='multinomial').fit(X_train_normalized, y_train)
```

```python

# Scaling numeric Test data
X_test_num = X_test.select_dtypes(include = np.number)

X_test_normalized = transformer.transform(X_test_num)
X_test_normalized = pd.DataFrame(X_test_normalized,columns=X_test_num.columns)
X_test_normalized.head()
```

```python


# Predictions on the test set:
predictions = classification.predict(X_test_normalized)
classification.score(X_test_normalized, y_test)
```

```python

# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
y_train.value_counts()
```

```python

# Model Evaluation Metrics

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

pred = classification.predict(X_test_normalized)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=100, k_neighbors=3)
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_train_normalized,y_train)
```

```python
X_train_SMOTE.shape
```

```python
LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_SMOTE, y_train_SMOTE)
pred = LR.predict(X_test_normalized)

print("classification score: ",classification.score(X_test_normalized, y_test))
print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```
